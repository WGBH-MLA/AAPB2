```python
from cmless.parse import parse_collections
```

```python
collections = parse_collections('../app/views/special_collections')
len(collections)
collections[0]
```

```python
with open('../collections.json', 'w') as f:
    import json

    json.dump([c.model_dump(by_alias=True) for c in collections], f, indent=2)
```
